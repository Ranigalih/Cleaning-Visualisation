```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
ax = sns.violinplot(x="sex", y="tip", inner="box", data=tips)
ax.set_title("Distribution of tips", fontsize=16);
```

```python
ax = sns.violinplot(x="sex", y="tip", inner="quartile", data=tips)
ax.set_title("Distribution of tips", fontsize=16);
```

```python
ax = sns.violinplot(x="sex", y="tip", inner="point", data=tips)
ax.set_title("Distribution of tips", fontsize=16);
```

```python
ax = sns.violinplot(x="sex", y="total_bill", inner="box", data=tips)
ax.set_title("Total Bill", fontsize=16);
```
