```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv("C:/Users/USER/Downloads/dataset_final_project.csv")
dataset
```

```python
# no 1

dataset["Branch"] = dataset["Branch"].replace(["A", "B" , "C"], [1,2,3])
dataset["City"] = dataset["City"].replace(["Yangoon", "Naypytaw","Mandalay"],[1,2,3])
dataset["Customer type"] = dataset["Customer type"].replace(["Member","Normal"], [1,2])
dataset["Product line"] = dataset["Product line"].replace(["Electronic accessories","Fashion accessories","Food and beverages", 
                                                           "Health and beauty", "Home and lifestyle", "Sports and travel"],[1,2,3,4,5,6])
dataset["Payment"] = dataset["Payment"].replace(["Cash","Credit card", "Ewallet"],[1,2,3])
dataset
```

```python
#no 2
dataset["Total"] = dataset["Unit price"] * dataset["Quantity"]
dataset["Tax"] = dataset["Total"] * 0.5
dataset["Final price"] = dataset["Total"] * dataset["Tax"]
dataset
```

```python
#no 3

dataset.dropna()
```

```python
#no 4

#a) 

jumlah = dataset.groupby("Product line")["Quantity"].sum()

plt.figure(figsize=[5,5])
jumlah.plot(kind="bar" , label="Quantity")
plt.grid(False)
plt.legend()
```

```python
#b

normal = dataset[dataset["Customer type"] == 1]
member = dataset[dataset["Customer type"] == 2]

product1 = normal.groupby("Product line")["Product line"].count()
product2 = member.groupby("Product line")["Product line"].count()

plt.figure(figsize=[7,5])
product1.plot(kind ="bar" , label = "Customer type")
plt.grid(False)
plt.legend()

plt.figure(figsize=[7,5])
product2.plot(kind= "bar" , label = "Customer type")
plt.grid(False)
plt.legend()

```
